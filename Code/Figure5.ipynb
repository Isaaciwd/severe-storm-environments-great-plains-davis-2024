## This notebook contains all the code necessary to recreate Figure 5

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

# Define the path to the datasets that store the figure data
stem = "../Data/"
```

# This cell is used to recreate Figure 5a. 
The data is saved in a nc file called Figure5a5c_data.nc

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure5a5c_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and Static Energy data
ERA5_z = ds['ERA5_geopotential_height'].values
CNRM_hist_z = ds['CNRM-Hist_geopotential_height'].values
MPI_hist_z = ds['MPI-Hist_geopotential_height'].values
ERA5_MSE = ds['ERA5_MSE'].values
CNRM_hist_MSE = ds['CNRM-Hist_MSE'].values
MPI_hist_MSE = ds['MPI-Hist_MSE'].values
ERA5_DSE = ds['ERA5_DSE'].values
CNRM_hist_DSE = ds['CNRM-Hist_DSE'].values
MPI_hist_DSE = ds['MPI-Hist_DSE'].values

# Extract difference data
mpi_MSE_diff = ds['MPI-hist_ERA5_MSE_diff'] 
cnrm_MSE_diff = ds['CNRM-hist_ERA5_MSE_diff']
mpi_DSE_diff = ds['MPI-hist_ERA5_DSE_diff']
cnrm_DSE_diff = ds['CNRM-hist_ERA5_DSE_diff']

# Plot MSE profiles and their percentiles
axs[0].plot(ERA5_MSE, ERA5_z, color='black', label='ERA5 Moist')
axs[0].plot(ERA5_DSE, ERA5_z, color='black', label='ERA5 Dry Dry', linestyle="--")
axs[0].plot(MPI_hist_MSE, MPI_hist_z, color='red', label='MPI-Hist Moist')
axs[0].plot(MPI_hist_DSE, MPI_hist_z, color='red', label='MPI-Hist Dry', linestyle="--")
axs[0].plot(CNRM_hist_MSE, CNRM_hist_z, color='blue', label='CNRM-Hist Moist')
axs[0].plot(CNRM_hist_DSE, CNRM_hist_z, color='blue', label='CNRM-Hist Dry', linestyle="--")


# Plot SE differences from ERA5
axs[1].plot(cnrm_MSE_diff, ERA5_z, color='blue', label='CNRM-ERA5 Moist')
axs[1].plot(cnrm_DSE_diff, ERA5_z, color='blue', label='CNRM-ERA5 Dry', linestyle="--")
axs[1].plot(mpi_MSE_diff, ERA5_z, color='red', label='MPI-ERA5 Moist')
axs[1].plot(mpi_DSE_diff, ERA5_z, color='red', label='MPI-ERA5 Dry', linestyle="--")
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center
axs[0].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center


# Set titles, labels, and limits
fig.suptitle('Historic Static Energy Profiles', x=0.4, y=0.94)
axs[0].set_ylim(0, 16)
axs[0].set_xlabel('Static Energy (kJ)')
axs[0].set_ylabel('Height Above Surface (km)')
axs[0].legend()
axs[0].set_xlim(300,389)

axs[1].set_ylim(0, 16)
axs[1].set_xlabel('Diff in Static Energy (KJ)')
axs[1].set_xlim(-20,20)
axs[1].legend(fontsize=6)
```

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure5b5d_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and MSE_deficit data
ERA5_z = ds['ERA5_geopotential_height'].values
CNRM_hist_z = ds['CNRM-Hist_geopotential_height'].values
MPI_hist_z = ds['MPI-Hist_geopotential_height'].values
ERA5_MSE_deficit = ds['ERA5_MSE_deficit'].values
CNRM_hist_MSE_deficit = ds['CNRM-Hist_MSE_deficit'].values
MPI_hist_MSE_deficit = ds['MPI-Hist_MSE_deficit'].values

# Extract 25th and 75th percentiles for MSE_deficit
ERA5_p25 = ds['ERA5_25th_Percentile_of_MSE_deficit'].values
ERA5_p75 = ds['ERA5_75th_Percentile_of_MSE_deficit'].values
CNRM_hist_p25 = ds['CNRM-Hist_25th_Percentile_of_MSE_deficit'].values
CNRM_hist_p75 = ds['CNRM-Hist_75th_Percentile_of_MSE_deficit'].values
MPI_hist_p25 = ds['MPI-Hist_25th_Percentile_of_MSE_deficit'].values
MPI_hist_p75 = ds['MPI-Hist_75th_Percentile_of_MSE_deficit'].values

# Extract difference data
mpi_diff = ds['MPI-hist_ERA5_MSE_deficit_diff'] 
cnrm_diff = ds['CNRM-hist_ERA5_MSE_deficit_diff']

# Plot MSE_deficit profiles and their percentiles
axs[0].plot(ERA5_MSE_deficit, ERA5_z, color='black', label='ERA5')
axs[0].plot(CNRM_hist_MSE_deficit, CNRM_hist_z, color='blue', label='CNRM-Hist')
axs[0].plot(MPI_hist_MSE_deficit, MPI_hist_z, color='red', label='MPI-Hist')
axs[0].fill_betweenx(ERA5_z, ERA5_p25, ERA5_p75, color='black', alpha=0.2)
axs[0].fill_betweenx(CNRM_hist_z, CNRM_hist_p25, CNRM_hist_p75, color='blue', alpha=0.2)
axs[0].fill_betweenx(MPI_hist_z, MPI_hist_p25, MPI_hist_p75, color='red', alpha=0.2)

# Plot MSE_deficit differences from ERA5
axs[1].plot(cnrm_diff, ERA5_z, color='blue', label='CNRM-ERA5')
axs[1].plot(mpi_diff, ERA5_z, color='red', label='MPI-ERA5')
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center
axs[0].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center


# Set titles, labels, and limits
fig.suptitle('Historic MSE Deficit Profiles', x=0.4, y=0.94)
axs[0].set_ylim(0, 16)
axs[0].set_xlabel('MSE Deficit (kJ)')
axs[0].set_ylabel('Height Above Surface (km)')
axs[0].legend()
axs[0].set_xlim(-40, 30)

axs[1].set_ylim(0, 16)
axs[1].set_xlabel('Diff in MSE Deficit (kJ)')
axs[1].set_xlim(-13,9)
axs[1].legend()
```

# This cell is used to recreate Figure 5c.

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure5a5c_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and Static Energy data
CNRM_hist_z = ds['CNRM-Hist_geopotential_height'].values
MPI_hist_z = ds['MPI-Hist_geopotential_height'].values
ERA5_MSE = ds['ERA5_MSE'].values
CNRM_hist_MSE = ds['CNRM-Hist_MSE'].values
MPI_hist_MSE = ds['MPI-Hist_MSE'].values
CNRM_hist_DSE = ds['CNRM-Hist_DSE'].values
MPI_hist_DSE = ds['MPI-Hist_DSE'].values

CNRM_fut_z = ds['CNRM-Fut_geopotential_height'].values
MPI_fut_z = ds['MPI-Fut_geopotential_height'].values
CNRM_fut_MSE = ds['CNRM-Fut_MSE'].values
MPI_fut_MSE = ds['MPI-Fut_MSE'].values
CNRM_fut_DSE = ds['CNRM-Fut_DSE'].values
MPI_fut_DSE = ds['MPI-Fut_DSE'].values

# Calculate difference data
mpi_MSE_diff = MPI_fut_MSE - MPI_hist_MSE
cnrm_MSE_diff = CNRM_fut_MSE - CNRM_hist_MSE
mpi_DSE_diff = MPI_fut_DSE - MPI_hist_DSE
cnrm_DSE_diff = CNRM_fut_DSE - CNRM_hist_DSE

# Plot MSE profiles and their percentiles
axs[0].plot(MPI_hist_MSE, MPI_hist_z, color='red', label='MPI-Hist Moist')
axs[0].plot(MPI_hist_DSE, MPI_hist_z, color='red', label='MPI-Hist Dry', linestyle="--")
axs[0].plot(MPI_fut_MSE, MPI_fut_z, color='coral', label='MPI-Fut Moist')
axs[0].plot(MPI_fut_DSE, MPI_fut_z, color='coral', label='MPI-Fut Dry', linestyle="--")


axs[0].plot(CNRM_hist_MSE, CNRM_hist_z, color='blue', label='CNRM-Hist Moist')
axs[0].plot(CNRM_hist_DSE, CNRM_hist_z, color='blue', label='CNRM-Hist Dry', linestyle="--")
axs[0].plot(CNRM_fut_MSE, CNRM_fut_z, color='cornflowerblue', label='CNRM-Fut Moist')
axs[0].plot(CNRM_fut_DSE, CNRM_fut_z, color='cornflowerblue', label='CNRM-Fut Dry', linestyle="--")


# Plot SE differences from ERA5
axs[1].plot(cnrm_MSE_diff, CNRM_hist_z, color='blue', label='CNRM-ERA5 Moist')
axs[1].plot(cnrm_DSE_diff, CNRM_hist_z, color='blue', label='CNRM-ERA5 Dry', linestyle="--")
axs[1].plot(mpi_MSE_diff, MPI_hist_z, color='red', label='MPI-ERA5 Moist')
axs[1].plot(mpi_DSE_diff, MPI_hist_z, color='red', label='MPI-ERA5 Dry', linestyle="--")
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center
axs[0].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center


# Set titles, labels, and limits
fig.suptitle('Future and Historic Static Energy Profiles', x=0.4, y=0.94)
axs[0].set_ylim(0, 16)
axs[0].set_xlabel('Static Energy (kJ)')
axs[0].set_ylabel('Height Above Surface (km)')
axs[0].legend()
axs[0].set_xlim(300,389)

axs[1].set_ylim(0, 16)
axs[1].set_xlabel('Diff in Static Energy (kJ)')
axs[1].set_xlim(-20,20)
axs[1].legend(fontsize=8)
```

# This cell is used to recreate Figure 5d. 
The data is saved in a nc file called Figure5b5d_data.nc

```python
# Load the climate data from a NetCDF file
ds = xr.open_dataset(f"{stem}/Figure5b5d_data.nc")

# Set up the figure and subplots
plt.rcParams["figure.figsize"] = [12, 5]  # Set figure size
fig, axs = plt.subplots(1, 2, sharey=True)  # Create subplots with shared y-axis
fig.subplots_adjust(wspace=0)  # Remove horizontal space between axes

# Adjust the size of the second subplot to make the difference graph smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and MSE_deficit data for historical and future scenarios
CNRM_hist_z = ds['CNRM-Hist_geopotential_height'].values
CNRM_fut_z = ds['CNRM-Fut_geopotential_height'].values
MPI_hist_z = ds['MPI-Hist_geopotential_height'].values
MPI_fut_z = ds['MPI-Fut_geopotential_height'].values

CNRM_hist_MSE_deficit = ds['CNRM-Hist_MSE_deficit'].values
CNRM_fut_MSE_deficit = ds['CNRM-Fut_MSE_deficit'].values
MPI_hist_MSE_deficit = ds['MPI-Hist_MSE_deficit'].values
MPI_fut_MSE_deficit = ds['MPI-Fut_MSE_deficit'].values

# Extract 25th and 75th percentiles of MSE_deficit
CNRM_hist_p25 = ds['CNRM-Hist_25th_Percentile_of_MSE_deficit'].values
CNRM_hist_p75 = ds['CNRM-Hist_75th_Percentile_of_MSE_deficit'].values
CNRM_fut_p25 = ds['CNRM-Fut_25th_Percentile_of_MSE_deficit'].values
CNRM_fut_p75 = ds['CNRM-Fut_75th_Percentile_of_MSE_deficit'].values
MPI_hist_p25 = ds['MPI-Hist_25th_Percentile_of_MSE_deficit'].values
MPI_hist_p75 = ds['MPI-Hist_75th_Percentile_of_MSE_deficit'].values
MPI_fut_p25 = ds['MPI-Fut_25th_Percentile_of_MSE_deficit'].values
MPI_fut_p75 = ds['MPI-Fut_75th_Percentile_of_MSE_deficit'].values

# Exctract p-values
MPI_MSE_deficit_pvalues = ds['MPI_MSE_deficit_pvalues'].values
CNRM_MSE_deficit_pvalues = ds['CNRM_MSE_deficit_pvalues'].values

# Plot historical and future MSE_deficit profiles for CNRM and MPI models
axs[0].plot(CNRM_hist_MSE_deficit, CNRM_hist_z, color='blue', label='CNRM-Hist')
axs[0].plot(CNRM_fut_MSE_deficit, CNRM_fut_z, color='cornflowerblue', label='CNRM-Fut', linestyle='--')
axs[0].plot(MPI_hist_MSE_deficit, MPI_hist_z, color='red', label='MPI-Hist')
axs[0].plot(MPI_fut_MSE_deficit, MPI_fut_z, color='coral', label='MPI-Fut', linestyle='--')

# Add shaded areas to represent the range between the 25th and 75th percentiles
axs[0].fill_betweenx(CNRM_hist_z, CNRM_hist_p25, CNRM_hist_p75, color='blue', alpha=0.2)
axs[0].fill_betweenx(CNRM_fut_z, CNRM_fut_p25, CNRM_fut_p75, color='cornflowerblue', alpha=0.2)
axs[0].fill_betweenx(MPI_hist_z, MPI_hist_p25, MPI_hist_p75, color='red', alpha=0.2)
axs[0].fill_betweenx(MPI_fut_z, MPI_fut_p25, MPI_fut_p75, color='coral', alpha=0.2)

# Calculate and plot the difference in MSE_deficit between future and historical scenarios
cnrm_diff = CNRM_fut_MSE_deficit - CNRM_hist_MSE_deficit
mpi_diff = MPI_fut_MSE_deficit - MPI_hist_MSE_deficit
axs[1].plot(cnrm_diff, CNRM_hist_z, color='blue', label='CNRM, Fut-Hist')
axs[1].plot(mpi_diff, MPI_hist_z, color='red', label='MPI, Fut-Hist')
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Add a dashed black line at 0
axs[0].plot([0, 0], [0, 16], linestyle='--', color='black')  # Add a dashed black line at 0


# Boolean arrays for condition
# Plot filled circles where p-value is less than 0.05, and unfilled in circles where p-value is greater than 0.05
less_than_005 = MPI_MSE_deficit_pvalues < 0.05
greater_or_equal_005 = MPI_MSE_deficit_pvalues >= 0.05
axs[1].scatter(mpi_diff[less_than_005], MPI_hist_z[less_than_005], color='red')
axs[1].scatter(mpi_diff[greater_or_equal_005], MPI_hist_z[greater_or_equal_005], facecolors='None', edgecolors='red')
less_than_005 = CNRM_MSE_deficit_pvalues < 0.05
greater_or_equal_005 = CNRM_MSE_deficit_pvalues >= 0.05
axs[1].scatter(cnrm_diff[less_than_005], CNRM_hist_z[less_than_005], color='blue')
axs[1].scatter(cnrm_diff[greater_or_equal_005], CNRM_hist_z[greater_or_equal_005], facecolors='None', edgecolors='blue')

# Configure plot aesthetics
fig.suptitle('Future and Historic MSE Deficit Profiles', x=0.4, y=0.93)
axs[0].set_ylim(0, 16)
axs[0].set_xlabel('MSE deficit (kJ)')
axs[0].set_ylabel('Height Above Surface (km)')
axs[0].set_xlim(-40, 30)
axs[0].legend()

axs[1].set_ylim(0, 16)
axs[1].set_xlabel('Dif in MSE deficit (kJ)')
axs[1].set_xlim(-13, 9)
axs[1].legend()
```
