# This cell is used to recreate Figure 3a. 
The data is saved in a nc file called Figure3_data.nc

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure3_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and temperature data
ERA5_z = ds['ERA5_geopotential_height']
CNRM_hist_z = ds['CNRM-Hist_geopotential_height']
MPI_hist_z = ds['MPI-Hist_geopotential_height']
ERA5_t = ds['ERA5_temperature']
CNRM_hist_t = ds['CNRM-Hist_temperature']
MPI_hist_t = ds['MPI-Hist_temperature']

# Extract 25th and 75th percentiles for temperature
ERA5_p25 = ds['ERA5_25th_Percentile_of_temperature']
ERA5_p75 = ds['ERA5_75th_Percentile_of_temperature']
CNRM_hist_p25 = ds['CNRM-Hist_25th_Percentile_of_temperature']
CNRM_hist_p75 = ds['CNRM-Hist_75th_Percentile_of_temperature']
MPI_hist_p25 = ds['MPI-Hist_25th_Percentile_of_temperature']
MPI_hist_p75 = ds['MPI-Hist_75th_Percentile_of_temperature']

#extract the data for the temperature differences
mpi_diff = ds['MPI-hist_ERA5_temperature_diff']
cnrm_diff = ds['CNRM-hist_ERA5_temperature_diff']

# Plot temperature profiles and their percentiles
axs[0].plot(ERA5_t, ERA5_z, color='black', label='ERA5')
axs[0].plot(CNRM_hist_t, CNRM_hist_z, color='blue', label='CNRM-Hist')
axs[0].plot(MPI_hist_t, MPI_hist_z, color='red', label='MPI-Hist')
axs[0].fill_betweenx(ERA5_z, ERA5_p25, ERA5_p75, color='black', alpha=0.2)
axs[0].fill_betweenx(CNRM_hist_z, CNRM_hist_p25, CNRM_hist_p75, color='blue', alpha=0.2)
axs[0].fill_betweenx(MPI_hist_z, MPI_hist_p25, MPI_hist_p75, color='red', alpha=0.2)

# Plot temperature differences from ERA5
axs[1].plot(cnrm_diff, ERA5_z, color='blue', label='CNRM-ERA5')
axs[1].plot(mpi_diff, ERA5_z, color='red', label='MPI-ERA5')
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center

# Set titles, labels, and limits
fig.suptitle('Historic Temperature Profiles', x=0.4, y=0.94)
axs[0].set_ylim(0, 16)
axs[0].set_xlabel('Temperature (K)')
axs[0].set_ylabel('Height Above Surface (km)')
axs[0].legend()
axs[0].set_xlim(200, 305)

axs[1].set_ylim(0, 16)
axs[1].set_xlabel('Diff in Temperature (K)')
axs[1].set_xlim(-7.9, 7.9)
axs[1].legend()
```

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure7c7f_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and MSE_deficit data
z = ds['geopotential_height'].values
CNRM_hist_shear = ds['CNRM-Hist_SRH'].values
MPI_hist_shear = ds['MPI-Hist_SRH'].values
CNRM_fut_shear = ds['CNRM-Fut_SRH'].values
MPI_fut_shear = ds['MPI-Fut_SRH'].values

# Extract 25th and 75th percentiles for MSE_deficit
CNRM_hist_p25 = ds['CNRM-Hist_25th_Percentile_of_SRH'].values
CNRM_hist_p75 = ds['CNRM-Hist_75th_Percentile_of_SRH'].values
MPI_hist_p25 = ds['MPI-Hist_25th_Percentile_of_SRH'].values
MPI_hist_p75 = ds['MPI-Hist_75th_Percentile_of_SRH'].values
CNRM_fut_p25 = ds['CNRM-Fut_25th_Percentile_of_SRH'].values
CNRM_fut_p75 = ds['CNRM-Fut_75th_Percentile_of_SRH'].values
MPI_fut_p25 = ds['MPI-Fut_25th_Percentile_of_SRH'].values
MPI_fut_p75 = ds['MPI-Fut_75th_Percentile_of_SRH'].values

# calculate percent difference between future and historic models
mpi_diff = (MPI_fut_shear - MPI_hist_shear) / MPI_hist_shear * 100
cnrm_diff = (CNRM_fut_shear - CNRM_hist_shear)  / CNRM_hist_shear * 100

# Load pvalues
MPI_pvalues = ds['MPI_SRH_pvalues'].values
CNRM_pvalues = ds['CNRM_SRH_pvalues'].values


# Plot MSE_deficit profiles and their percentiles
axs[0].plot(CNRM_hist_shear, z, color='blue', label='CNRM-Hist')
axs[0].plot(MPI_hist_shear, z, color='red', label='MPI-Hist')
axs[0].plot(CNRM_fut_shear, z, color='cornflowerblue', label='CNRM-Fut', linestyle='--')
axs[0].plot(MPI_fut_shear, z, color='coral', label='MPI-Fut', linestyle='--')
axs[0].fill_betweenx(z, CNRM_hist_p25, CNRM_hist_p75, color='blue', alpha=0.2)
axs[0].fill_betweenx(z, MPI_hist_p25, MPI_hist_p75, color='red', alpha=0.2)
axs[0].fill_betweenx(z, CNRM_fut_p25, CNRM_fut_p75, color='cornflowerblue', alpha=0.2)
axs[0].fill_betweenx(z, MPI_fut_p25, MPI_fut_p75, color='coral', alpha=0.2)


# Plot shear differences from ERA5
axs[1].plot(cnrm_diff, z, color='blue', label='CNRM, Fut vs Hist')
axs[1].plot(mpi_diff, z, color='red', label='MPI, Fut vs Hist')
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center

# Plot filled circles where p-value is less than 0.05, and unfilled in circles where p-value is greater than 0.05
less_than_005 = MPI_pvalues < 0.05
greater_or_equal_005 = MPI_pvalues >= 0.05
axs[1].scatter(mpi_diff[less_than_005], z[less_than_005], color='red')
axs[1].scatter(mpi_diff[greater_or_equal_005], z[greater_or_equal_005], facecolors='None', edgecolors='red')
less_than_005 = CNRM_pvalues < 0.05
greater_or_equal_005 = CNRM_pvalues >= 0.05
axs[1].scatter(cnrm_diff[less_than_005], z[less_than_005], color='blue')
axs[1].scatter(cnrm_diff[greater_or_equal_005], z[greater_or_equal_005], facecolors='None', edgecolors='blue')


# Set titles, labels, and limits
fig.suptitle('Future and Historic Storm Relative Helicity', x=0.4, y=0.94)
axs[0].set_ylabel('Layer-top Height Above Surface (km)', fontsize = fontsize-2)
axs[0].set_ylim(0,)
axs[0].legend()
axs[0].set_xlabel(f'SRH (m²/s²)', fontsize = fontsize-2)

axs[1].set_ylim(0.4,3) 
axs[1].set_xlabel('% Dif in SRH')
axs[1].legend()
```
