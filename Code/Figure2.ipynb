## This notebook contains all the code necessary to recreate Figure 2

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

# Define the path to the datasets that store the figure data
stem = "../Data/"
```

# This cell is used to recreate Figure 2a. 
The data is saved in a nc file called Figure2a_data.nc

```python
# Open the dataset using xarray
ds = xr.open_dataset(f"{stem}/Figure2a_data.nc")

# Extract the count data for each climate model 
ERA5count = ds['ERA5_count'].values
MPIhistcount = ds['MPI-Hist_count'].values
MPIfutcount = ds['MPI-Fut_count'].values
CNRMhistcount = ds['CNRM-Hist_count'].values
CNRMfutcount = ds['CNRM-Fut_count'].values

# Extract month names from the dataset
monthnames = ds.month.values

# Define x-axis positions for the bars
x = np.arange(1, 13, 1)  # Months 1 through 12
w = 0.19  # Width of the bars

# Plotting
# Plot bars for each dataset with specific color and label
plt.bar(x - 2*w, ERA5count, width=w, align='center', color='black', label='ERA5')
plt.bar(x - w, CNRMhistcount, width=w, align='center', color='blue', label='CNRM Hist')
plt.bar(x, CNRMfutcount, width=w, align='center', color='cornflowerblue', label='CNRM Fut')
plt.bar(x + w, MPIhistcount, width=w, align='center', color='red', label='MPI Hist')
plt.bar(x + 2*w, MPIfutcount, width=w, align='center', color='coral', label='MPI Fut')

# Set x-ticks to month names and rotate them for better readability
plt.xticks(x, monthnames, rotation=-24)

# Set plot title and labels for axes
plt.title('Number of Soundings vs Month')
plt.ylabel('Normalised Count')

# Display legend
plt.legend()
```

# This cell is used to recreate Figure 2b. 
The data is saved in a nc file called Figure2b_data.nc

```python
# open the dataset using xarray
ds = xr.open_dataset(f"{stem}/Figure2b_data.nc")

# Extract the count data for each climate model and observation dataset
ERA5count = ds['ERA5_diurnal_cycle_count'].values
MPIhistcount = ds['MPI-Hist_diurnal_cycle_count'].values
MPIfutcount = ds['MPI-Fut_diurnal_cycle_count'].values
CNRMhistcount = ds['CNRM-Hist_diurnal_cycle_count'].values
CNRMfutcount = ds['CNRM-Fut_diurnal_cycle_count'].values
hour = ds.hour.values

# Define x-axis positions for the bars
x = np.arange(0,24,6)          
w = 1

# Plotting
plt.bar(x-2*w, ERA5count[::6], width = w, align='center', color ='black', label = 'ERA5')
plt.bar(x+w, MPIhistcount[::6], width = w, align='center', color ='red', label = 'MPI Hist')
plt.bar(x+2*w, MPIfutcount[::6], width = w, align='center', color ='coral', label = 'MPI Fut')
plt.bar(x-w, CNRMhistcount[::6], width = w, align='center', color ='blue', label = 'CNRM Hist')
plt.bar(x, CNRMfutcount[::6], width = w, align='center', color ='cornflowerblue', label = 'CNRM Fut')

# Set plot title and labels for axes
plt.xlabel("Hour UTC", x = 0.45)
plt.ylabel("Normalised Count")
plt.title("Number of Soundings vs Hour UTC")

# Display legend
plt.legend()
```

# This cell is used to recreate Figure 2c. 
The data is saved in a nc file called Figure2c_data.nc

```python
# Open the dataset using xarray
ds = xr.open_dataset(f"{stem}/Figure2c_data.nc")

# Create an empty list to store the data
cin_histo_data = []

# Extract the count data for each climate model and observation dataset
ERA5cin = ds['ERA5_CIN'].values
MPIhistcin = ds['MPI-Hist_CIN'].values
MPIfutcin = ds['MPI-Fut_CIN'].values
CNRMhistcin = ds['CNRM-Hist_CIN'].values
CNRMfutcin = ds['CNRM-Fut_CIN'].values

# Append the data to the list
cin_histo_data.append(ERA5cin)
cin_histo_data.append(MPIhistcin)
cin_histo_data.append(MPIfutcin)
cin_histo_data.append(CNRMhistcin)
cin_histo_data.append(CNRMfutcin)

# define histogram bins
bins = np.arange(0,121,20)

# Plotting
labels = ["ERA5", 'MPI Hist', "MPI Fut", 'CNRM Hist', "CNRM Fut"]
colors = ['k', 'r', 'coral', 'b', 'cornflowerblue']
plt.hist(cin_histo_data, bins, color = colors, label=labels)

# Set plot title and labels for axes
plt.title('Convective Inhibition Histogram')
plt.ylabel('Count')
plt.xlabel('Convective Inhibition (J kg)')
plt.legend()
```

# This cell is used to recreate Figure 2d. 
The data is saved in a nc file called Figure2d_data.nc

```python
# Open the dataset using xarray
ds = xr.open_dataset(f"{stem}/Figure2d_data.nc")

# Create an empty list to store the data
LCL_histo_data = []

# Extract the count data for each climate model and observation dataset
ERA5LCL = ds['ERA5_LCL'].values
MPIhistLCL = ds['MPI-Hist_LCL'].values
MPIfutLCL = ds['MPI-Fut_LCL'].values
CNRMhistLCL = ds['CNRM-Hist_LCL'].values
CNRMfutLCL = ds['CNRM-Fut_LCL'].values

# Append the data to the list
LCL_histo_data.append(ERA5LCL)
LCL_histo_data.append(MPIhistLCL)
LCL_histo_data.append(MPIfutLCL)
LCL_histo_data.append(CNRMhistLCL)
LCL_histo_data.append(CNRMfutLCL)

# define histogram bins
bins =  np.arange(0,4000,500)

# Plotting
labels = ["ERA5", 'MPI Hist', "MPI Fut", 'CNRM Hist', "CNRM Fut"]
colors = ['k', 'r', 'coral', 'b', 'cornflowerblue']
plt.hist(LCL_histo_data, bins, color = colors, label=labels)

# Set plot title and labels for axes
plt.title('Lifted Condensation Level Histogram')
plt.ylabel('Count')
plt.xlabel('Meters above ground level')
plt.legend()
```
