## This notebook contains all the code necessary to recreate Figure 6

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from metpy.plots import Hodograph


# Define the path to the datasets that store the figure data
stem = "../Data/"
```

# This cell is used to recreate Figure 6a. 
The data is saved in a nc file called Figure6a6c_data.nc

```python
from metpy.plots import Hodograph

# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure6a6c_data.nc")

# Make a figure and axis
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
# Create a hodograph
h = Hodograph(ax, component_range = 25)
# Add "range rings" to the plot
h.add_grid(increment=5)

# Extract geopotential height and Static Energy data
ERA5_z = ds['ERA5_geopotential_height'].values
CNRM_hist_z = ds['CNRM-Hist_geopotential_height'].values
MPI_hist_z = ds['MPI-Hist_geopotential_height'].values
ERA5_U = ds['ERA5_U'].values
CNRM_hist_U = ds['CNRM-Hist_U'].values
MPI_hist_U = ds['MPI-Hist_U'].values
ERA5_V = ds['ERA5_V'].values
CNRM_hist_V = ds['CNRM-Hist_V'].values
MPI_hist_V = ds['MPI-Hist_V'].values

# Load storm motion data
ERA5_sm = ds['ERA5_storm_motion'].values
CNRM_hist_sm = ds['CNRM-Hist_storm_motion'].values
MPI_hist_sm = ds['MPI-Hist_storm_motion'].values


# Plot the wind data
h.plot(ERA5_U, ERA5_V, c = 'k', label = "ERA5" , zorder = 2)
h.plot(MPI_hist_U, MPI_hist_V, c = 'r', label = "MPI-Hist" , zorder = 2)
h.plot(CNRM_hist_U, CNRM_hist_V, c = 'b', label = "CNRM-Hist" , zorder = 2)

# Plot the storm motion 
ax.text(ERA5_sm[0], ERA5_sm[1], 'M', style ='italic', color= 'k')
ax.text(MPI_hist_sm[0], MPI_hist_sm[1], 'M', style ='italic', color= 'r')
ax.text(CNRM_hist_sm[0], CNRM_hist_sm[1], 'M', style ='italic', color= 'b')

# create legend in top left
ax.legend(loc='upper left')

#Labeling graph
fontsize = 15
ax.set_xlabel('m/s', fontsize = fontsize)
ax.set_ylabel('m/s', fontsize = fontsize)
ax.set_xlim(-4,23)
ax.set_ylim(-7,15)

# Add a title
fig.suptitle('Hodograph of Historic Wind Profiles', y=0.85 )
```
