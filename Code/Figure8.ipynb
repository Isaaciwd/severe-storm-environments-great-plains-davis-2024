## This notebook contains all the code necessary to recreate Figure 8

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

# Define the path to the datasets that store the figure data
stem = "../Data/"
```

# This cell is used to recreate Figure 8a. 
The data is saved in a nc file called Figure8a_data.nc

```python
# Load the climate data from a NetCDF file
ds = xr.open_dataset(f"{stem}/Figure8a_data.nc")

# Set up the figure and subplots
plt.rcParams["figure.figsize"] = [12, 5]  # Set figure size
fig, axs = plt.subplots()  # Create plot


# Extract geopotential height and RH data for historical and future scenarios
CNRM_hist_z = ds['CNRM-Hist_geopotential_height'].values
CNRM_fut_z = ds['CNRM-Fut_geopotential_height'].values
MPI_hist_z = ds['MPI-Hist_geopotential_height'].values
MPI_fut_z = ds['MPI-Fut_geopotential_height'].values
ERA5_z = ds['ERA5_geopotential_height'].values


CNRM_hist_RH = ds['CNRM-Hist_RH_interannual_variability'].values
CNRM_fut_RH = ds['CNRM-Fut_RH_interannual_variability'].values
MPI_hist_RH = ds['MPI-Hist_RH_interannual_variability'].values
MPI_fut_RH = ds['MPI-Fut_RH_interannual_variability'].values
ERA5_RH = ds['ERA5_RH_interannual_variability'].values


# Plot historical and future RH profiles for CNRM and MPI models
axs.plot(ERA5_RH, ERA5_z, color='black', label='ERA5')
axs.plot(CNRM_hist_RH, CNRM_hist_z, color='blue', label='CNRM-Hist')
axs.plot(CNRM_fut_RH, CNRM_fut_z, color='cornflowerblue', label='CNRM-Fut', linestyle='--')
axs.plot(MPI_hist_RH, MPI_hist_z, color='red', label='MPI-Hist')
axs.plot(MPI_fut_RH, MPI_fut_z, color='coral', label='MPI-Fut', linestyle='--')


# Configure plot aesthetics
fig.suptitle('Relative Humidity Interannual Range',y=0.93)
axs.set_ylim(0, 16)
axs.set_xlabel('Interannual Range of RH (%)')
axs.set_ylabel('Height Above Surface (km)')
axs.legend()
```
