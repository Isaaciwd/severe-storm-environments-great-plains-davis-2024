## This notebook contains all the code necessary to recreate Figure 7

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

# Define the path to the datasets that store the figure data
stem = "../Data/"
```

# This cell is used to recreate Figure 7a. 
The data is saved in a nc file called Figure7a7d_data.nc

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure7a7d_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and MSE_deficit data
z = ds['geopotential_height'].values
ERA5_shear = ds['ERA5_shear'].values
CNRM_hist_shear = ds['CNRM-Hist_shear'].values
MPI_hist_shear = ds['MPI-Hist_shear'].values

# Extract 25th and 75th percentiles for MSE_deficit
ERA5_p25 = ds['ERA5_25th_Percentile_of_shear'].values
ERA5_p75 = ds['ERA5_75th_Percentile_of_shear'].values
CNRM_hist_p25 = ds['CNRM-Hist_25th_Percentile_of_shear'].values
CNRM_hist_p75 = ds['CNRM-Hist_75th_Percentile_of_shear'].values
MPI_hist_p25 = ds['MPI-Hist_25th_Percentile_of_shear'].values
MPI_hist_p75 = ds['MPI-Hist_75th_Percentile_of_shear'].values

# calculate percent difference between models and ERA5
mpi_diff = (MPI_hist_shear - ERA5_shear) / ERA5_shear * 100
cnrm_diff = (CNRM_hist_shear - ERA5_shear)  / ERA5_shear * 100

# Plot MSE_deficit profiles and their percentiles
axs[0].plot(ERA5_shear, z, color='black', label='ERA5')
axs[0].plot(CNRM_hist_shear, z, color='blue', label='CNRM-Hist')
axs[0].plot(MPI_hist_shear, z, color='red', label='MPI-Hist')
axs[0].fill_betweenx(z, ERA5_p25, ERA5_p75, color='black', alpha=0.2)
axs[0].fill_betweenx(z, CNRM_hist_p25, CNRM_hist_p75, color='blue', alpha=0.2)
axs[0].fill_betweenx(z, MPI_hist_p25, MPI_hist_p75, color='red', alpha=0.2)

# Plot shear differences from ERA5
axs[1].plot(cnrm_diff, z, color='blue', label='CNRM-Hist vs ERA5')
axs[1].plot(mpi_diff, z, color='red', label='MPI-Hist vs ERA5')
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center


# Set titles, labels, and limits
fontsize = 15
fig.suptitle('Historic Integrated Bulk Shear', x=0.4, y=0.94)
axs[0].set_ylabel('Layer-top Height Above Surface (km)', fontsize = fontsize-2)
axs[0].set_ylim(0,)
axs[0].legend()
axs[0].set_xlabel(f'Wind Shear (m/s)', fontsize = fontsize-2)

axs[1].set_ylim(0,6) 
axs[1].set_xlabel('% Dif in Windshear')
axs[1].legend()
```

```python
# Load the dataset
ds = xr.open_dataset(f"{stem}/Figure7a7d_data.nc")

# Set the figure size and create subplots with shared y-axis
plt.rcParams["figure.figsize"] = [12, 5]
fig, axs = plt.subplots(1, 2, sharey=True)

# Adjust the space between the subplots
fig.subplots_adjust(wspace=0)

# Adjust the size of the second subplot to make it smaller
chartBox = axs[1].get_position()
axs[1].set_position([chartBox.x0, chartBox.y0, chartBox.width * 0.6, chartBox.height])

# Extract geopotential height and MSE_deficit data
z = ds['geopotential_height'].values
CNRM_hist_shear = ds['CNRM-Hist_shear'].values
MPI_hist_shear = ds['MPI-Hist_shear'].values
CNRM_fut_shear = ds['CNRM-Fut_shear'].values
MPI_fut_shear = ds['MPI-Fut_shear'].values

# Extract 25th and 75th percentiles for MSE_deficit
CNRM_hist_p25 = ds['CNRM-Hist_25th_Percentile_of_shear'].values
CNRM_hist_p75 = ds['CNRM-Hist_75th_Percentile_of_shear'].values
MPI_hist_p25 = ds['MPI-Hist_25th_Percentile_of_shear'].values
MPI_hist_p75 = ds['MPI-Hist_75th_Percentile_of_shear'].values
CNRM_fut_p25 = ds['CNRM-Fut_25th_Percentile_of_shear'].values
CNRM_fut_p75 = ds['CNRM-Fut_75th_Percentile_of_shear'].values
MPI_fut_p25 = ds['MPI-Fut_25th_Percentile_of_shear'].values
MPI_fut_p75 = ds['MPI-Fut_75th_Percentile_of_shear'].values

# calculate percent difference between future and historic models
mpi_diff = (MPI_fut_shear - MPI_hist_shear) / MPI_hist_shear * 100
cnrm_diff = (CNRM_fut_shear - CNRM_hist_shear)  / CNRM_hist_shear * 100

# Load pvalues
MPI_pvalues = ds['MPI_shear_pvalues'].values
CNRM_pvalues = ds['CNRM_shear_pvalues'].values


# Plot MSE_deficit profiles and their percentiles
axs[0].plot(CNRM_hist_shear, z, color='blue', label='CNRM-Hist')
axs[0].plot(MPI_hist_shear, z, color='red', label='MPI-Hist')
axs[0].plot(CNRM_fut_shear, z, color='cornflowerblue', label='CNRM-Fut', linestyle='--')
axs[0].plot(MPI_fut_shear, z, color='coral', label='MPI-Fut', linestyle='--')
axs[0].fill_betweenx(z, CNRM_hist_p25, CNRM_hist_p75, color='blue', alpha=0.2)
axs[0].fill_betweenx(z, MPI_hist_p25, MPI_hist_p75, color='red', alpha=0.2)
axs[0].fill_betweenx(z, CNRM_fut_p25, CNRM_fut_p75, color='cornflowerblue', alpha=0.2)
axs[0].fill_betweenx(z, MPI_fut_p25, MPI_fut_p75, color='coral', alpha=0.2)


# Plot shear differences from ERA5
axs[1].plot(cnrm_diff, z, color='blue', label='CNRM, Fut vs Hist')
axs[1].plot(mpi_diff, z, color='red', label='MPI, Fut vs Hist')
axs[1].plot([0, 0], [0, 16], linestyle='--', color='black')  # Dashed black line at the center

# Plot filled circles where p-value is less than 0.05, and unfilled in circles where p-value is greater than 0.05
less_than_005 = MPI_pvalues < 0.05
greater_or_equal_005 = MPI_pvalues >= 0.05
axs[1].scatter(mpi_diff[less_than_005], z[less_than_005], color='red')
axs[1].scatter(mpi_diff[greater_or_equal_005], z[greater_or_equal_005], facecolors='None', edgecolors='red')
less_than_005 = CNRM_pvalues < 0.05
greater_or_equal_005 = CNRM_pvalues >= 0.05
axs[1].scatter(cnrm_diff[less_than_005], z[less_than_005], color='blue')
axs[1].scatter(cnrm_diff[greater_or_equal_005], z[greater_or_equal_005], facecolors='None', edgecolors='blue')


# Set titles, labels, and limits
fig.suptitle('Future and Historic Integrated Bulk Shear', x=0.4, y=0.94)
axs[0].set_ylabel('Layer-top Height Above Surface (km)', fontsize = fontsize-2)
axs[0].set_ylim(0,)
axs[0].legend()
axs[0].set_xlabel(f'Wind Shear (m/s)', fontsize = fontsize-2)

axs[1].set_ylim(0,6) 
axs[1].set_xlabel('% Dif in Windshear')
axs[1].legend()
```
